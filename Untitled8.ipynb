```python
import numpy as np
outliers = []
def detect_outliers_zscore(ALP):
    thres = 3
    mean = np.mean(df.ALP)
    std = np.std(df.ALP)
    print(mean, std)
    for i in df.ALP:
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers# Driver code
df.ALP_outliers = detect_outliers_zscore(df.ALP)
print("Outliers from Z-scores method: ", df.ALP_outliers)
```

```python
outliers = []
def detect_outliers_iqr(data):
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    # print(q1, q3)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    # print(lwr_bound, upr_bound)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers# Driver code
df.Income_outliers = detect_outliers_iqr(df.ALP)
print("Outliers from IQR method: ", df.ALP_outliers)
```

```python
tenth_percentile = np.percentile(df.ALP, 10)
ninetieth_percentile = np.percentile(df.ALP, 90)
print(tenth_percentile, ninetieth_percentile) 
a = np.where(df.ALP<tenth_percentile, tenth_percentile, df.ALP)
a = np.where(a>ninetieth_percentile, ninetieth_percentile, a)
```

```python
plt.boxplot(a, vert=False)
plt.title("Boxplot of the sample after treating the outliers")
plt.xlabel("ALP")
```
